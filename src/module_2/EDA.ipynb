```python
from pathlib import Path
import pandas as pd
```

```python
def main_checks_dataframe_column(column : list):
    try:
        print(f"Total values: {column.count()}")
        print(f"Number of null values: {column.isnull().sum()} is the {column.isnull().mean() * 100 :.2f}% of data")
        print(f"Duplicates Values: {column.duplicated().sum()} is the {column.duplicated().mean() * 100 :.2f}% of data") #Number of null values: 51 is the 1.0234798314268514 of data
        print(f"Number of types: {len(column.apply(type).unique())} types: {column.apply(type).unique()}")
        if len(column.apply(type).unique()) == 1:
            if column.apply(type).unique() == float or column.apply(type).unique() == float:
                print(f"The mean: {column.mean()}")
                print(f"The standard deviation: {column.std()}")
                print(f"The min: {column.min()} The max: {column.max()} ")
        print(f"Number of values uniques: {len(column.unique())}")
        if len(column.unique()) < column.count()*0.50:
            print(f"The unique values : {column.unique() :.2f} ")
    except:
        print("Error")   
```

```python
# Ruta de la carpeta que contiene los archivos .parquet
folder_path = Path("c:/Users/AULA04/Desktop/Codes/zrive-ds/data/sampled-datasets")

# Diccionario para almacenar los DataFrames
sampled_datasets = {}

# Recorrer todos los archivos .parquet en la carpeta
for file_path in folder_path.glob("*.parquet"):  # Busca todos los archivos .parquet
    # Crear el nombre de la variable sin extensión
    file_name = file_path.stem  # Nombre del archivo sin extensión
    print(f"Leyendo archivo: {file_name}")
    
    # Leer el archivo .parquet y guardar en el diccionario
    sampled_datasets[file_name] = pd.read_parquet(file_path)


# Ahora puedes acceder a cada DataFrame por su nombre de archivo
print(f"Archivos cargados: {list(sampled_datasets.keys())}")

```
